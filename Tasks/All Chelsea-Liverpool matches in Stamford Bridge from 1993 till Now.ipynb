```python
from bs4 import BeautifulSoup 
import requests
```

```python
url = "https://en.wikipedia.org/wiki/Chelsea_F.C.%E2%80%93Liverpool_F.C._rivalry"
page_req = requests.get(url)
page_cont = page_req.content
```

```python
soup = BeautifulSoup(page_cont,'lxml')
```

```python
main_table = soup.find("div",{'class':'mw-content-ltr mw-parser-output'}).find_all('tr')

print("All Chelsea-Liverpool matches in Stamford Bridge from 1993 till Now")
print(40*'-')
for i in range(48,80):
    table = main_table[i].find_all('td')
    time = table[1].text.strip()
    result = table[2].text.strip()
    print(f'Chelsea {result} Liverpool \n  {time}')
    Chelsea_goalscorers = table[3].text.strip()
    Liverpool_goalscorers = table[4].text.strip()
    if Chelsea_goalscorers =="":
        pass
    else :
        print(f'Chelsea goalscorers : {Chelsea_goalscorers}')
    if Liverpool_goalscorers =="":
        pass
    else :
        print(f'Liverpool goalscorers : {Liverpool_goalscorers}')
    print(20*'-')
```
